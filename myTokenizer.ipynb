```python
import os
# 在导入 transformers 之前设置镜像！
os.environ['HF_ENDPOINT'] = 'https://hf-mirror.com'

from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("./gpt2-local")
# 2. 编码（文本→数字）
text = "大模型很强大！"
encoded = tokenizer.encode(text)
print(f"Token IDs: {encoded}")
# 输出: [1234, 5678, 9012, 3456]

# 3. 解码（数字→文本）
decoded = tokenizer.decode(encoded)
print(f"恢复文本: {decoded}")
# 输出: "大模型很强大！"

# 4. 查看token详情
tokens = tokenizer.convert_ids_to_tokens(encoded)
print(f"Tokens: {tokens}")
# 输出: ['大', '模', '型', '很', '强', '大', '！']

# 5. 获取词表大小
vocab_size = tokenizer.vocab_size
print(f"词表大小: {vocab_size}")
# GPT-2: 50257, LLaMA: 32000, GPT-4: ~100000
```

```python
def visualize_tokens(text, tokenizer):
    """可视化token切分效果"""
    tokens = tokenizer.tokenize(text)
    ids = tokenizer.convert_tokens_to_ids(tokens)
    
    print(f"原文: {text}")
    print(f"长度: {len(text)} 字符")
    print("-" * 50)
    for i, (token, id) in enumerate(zip(tokens, ids)):
        print(f"[{i}] '{token}' → ID: {id}")
    print("-" * 50)
    print(f"Token数量: {len(tokens)}")

# 使用示例
text = "ChatGPT is amazing!"
visualize_tokens(text, tokenizer)
```
