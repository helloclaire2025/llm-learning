```python
import evaluate

accuracy = evaluate.load("accuracy")

# 真实标签和预测标签
references = [0, 1, 1, 0, 1, 1, 0]
predictions = [0, 1, 0, 0, 1, 1, 1]

result = accuracy.compute(predictions=predictions, references=references)
print(result)  # {'accuracy': 0.714}
```

```python
f1 = evaluate.load("f1")
precision = evaluate.load("precision")
recall = evaluate.load("recall")

result_f1 = f1.compute(predictions=predictions, references=references)
result_precision = precision.compute(predictions=predictions, references=references)
result_recall = recall.compute(predictions=predictions, references=references)

print(result_f1)  # {'f1': 0.75}
print(result_precision)  # {'precision': 0.8}
print(result_recall)  # {'recall': 0.75}
```

```python
# 加载BLEU指标
bleu = evaluate.load("bleu")

# 准备数据
predictions = ["The cat is on the table"]
references = [["The cat is on the table", "A cat sits on the table"]]

# 计算BLEU分数
results = bleu.compute(predictions=predictions, references=references)
print(results)
print(f"BLEU Score: {results['bleu']}")
print(f"BLEU-1: {results['precisions'][0]}")
print(f"BLEU-2: {results['precisions'][1]}")
print(f"BLEU-3: {results['precisions'][2]}")
print(f"BLEU-4: {results['precisions'][3]}")
```

```python


# 加载ROUGE指标
rouge = evaluate.load("rouge")

# 准备数据
predictions = ["The cat sat on the mat"]
references = ["The cat is sitting on the mat"]

# 计算ROUGE分数
results = rouge.compute(predictions=predictions, references=references)
print(results)
print(f"ROUGE-1: {results['rouge1']}")
print(f"ROUGE-2: {results['rouge2']}")
print(f"ROUGE-L: {results['rougeL']}")
```

```python
# 加载指标
bleu = evaluate.load("bleu")
rouge = evaluate.load("rouge")

# 批量数据
predictions = [
    "The cat is on the table",
    "Machine translation is improving"
]
references_bleu = [
    ["The cat is on the table", "A cat sits on the table"],
    ["Machine translation keeps improving", "Translation quality is getting better"]
]
references_rouge = [
    "The cat is on the table",
    "Machine translation continues to improve"
]

# 批量计算
bleu_results = bleu.compute(predictions=predictions, references=references_bleu)
rouge_results = rouge.compute(predictions=predictions, references=references_rouge)

print(f"平均BLEU: {bleu_results['bleu']:.4f}")
print(f"平均ROUGE-L: {rouge_results['rougeL']:.4f}")
```

```python

from datasets import load_dataset

# 1. 加载数据集
dataset = load_dataset("cnn_dailymail", "3.0.0", split="test[:100]")

# 2. 加载评估指标
rouge = evaluate.load("rouge")
bleu = evaluate.load("bleu")

# 3. 准备预测和参考文本
predictions = dataset["highlights"][:50]  # 模型预测结果
references = dataset["article"][:50]      # 参考文本

# 4. 计算评估指标
rouge_results = rouge.compute(predictions=predictions, references=references)
bleu_results = bleu.compute(
    predictions=predictions, 
    references=[[ref] for ref in references]
)

# 5. 输出结果
print("=" * 50)
print("文本摘要评估结果")
print("=" * 50)
print(f"ROUGE-1: {rouge_results['rouge1']:.4f}")
print(f"ROUGE-2: {rouge_results['rouge2']:.4f}")
print(f"ROUGE-L: {rouge_results['rougeL']:.4f}")
print(f"BLEU-4: {bleu_results['bleu']:.4f}")
```
